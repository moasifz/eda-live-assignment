```python
#Q-10 What is the relationship between selling_price and km_driven for first-owner bikes?
#ANS-
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for first-owner bikes
first_owner_bikes = bike_data[bike_data['owner'] == 'First Owner']

# Plot the relationship between selling_price and km_driven
plt.figure(figsize=(10, 6))
sns.scatterplot(data=first_owner_bikes, x='km_driven', y='selling_price', alpha=0.7)
plt.title('Relationship between Selling Price and KM Driven for First-Owner Bikes')
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price (in currency units)')
plt.grid(True)
plt.show()
```

```python
#Q-11 Identify and remove outliers in the km_driven column using the IQR method?
#ANS
# Calculate Q1, Q3, and IQR for km_driven
Q1 = bike_data['km_driven'].quantile(0.25)
Q3 = bike_data['km_driven'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = bike_data[(bike_data['km_driven'] < lower_bound) | (bike_data['km_driven'] > upper_bound)]

# Remove outliers
bike_data_cleaned = bike_data[(bike_data['km_driven'] >= lower_bound) & (bike_data['km_driven'] <= upper_bound)]

# Output the number of removed outliers and the cleaned dataset shape
len(outliers), bike_data_cleaned.shape

```

```python
#Q-12 Perform a bivariate analysis to visualize the relationship between year and selling_price?
#ANS-
# Plot the relationship between year and selling_price
plt.figure(figsize=(12, 6))
sns.boxplot(data=bike_data, x='year', y='selling_price')
plt.title('Bivariate Analysis: Year vs Selling Price')
plt.xlabel('Year of Manufacture')
plt.ylabel('Selling Price (in currency units)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q-13What is the average depreciation in selling price based on the bike's age (current year - manufacturingyear?
#ANS-
# Add a new column for bike age
current_year = pd.Timestamp.now().year
bike_data['age'] = current_year - bike_data['year']

# Calculate the average selling price depreciation per year of age
average_depreciation = bike_data.groupby('age')['selling_price'].mean().diff().fillna(0)
average_depreciation
```

```python
#Q-14 Which bike names are priced significantly above the average price for their manufacturing year?
#ANS-
# Calculate the average selling price for each manufacturing year
average_price_by_year = bike_data.groupby('year')['selling_price'].mean()

# Identify bikes priced significantly above the average price (e.g., 1.5 times above the average)
threshold_multiplier = 1.5
bikes_above_avg = bike_data[bike_data.apply(
    lambda row: row['selling_price'] > threshold_multiplier * average_price_by_year[row['year']], axis=1)]

# Extract the bike names and their details
bikes_above_avg_names = bikes_above_avg[['name', 'year', 'selling_price']]
bikes_above_avg_names
```

```python
#Q-15 Develop a correlation matrix for numeric columns and visualize it using a heatmap?
#ANS-
# Calculate the correlation matrix for numeric columns
numeric_columns = bike_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = bike_data[numeric_columns].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix for Numeric Columns')
plt.show()
```

```python
#Q-1 What is the average selling price of cars for each dealer, and how does it compare across different dealers?
#ANS-
import pandas as pd

# Load the uploaded CSV file
file_path = '/mnt/data/Car Sale.csv'
car_data = pd.read_csv("Car Sale.csv")

# Display the first few rows to understand the structure of the data
car_data.head(), car_data.info()

```

```python
#Q-2 Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends?
#ANS-
price_variation_by_company = car_data.groupby('Company')['Price ($)'].std().sort_values(ascending=False)

# Find the company with the highest price variation
highest_variation_company = price_variation_by_company.idxmax()
highest_variation_value = price_variation_by_company.max()

# Display the results
highest_variation_company, highest_variation_value, price_variation_by_company


# Calculate the price variation (standard deviation) for each car brand (Company)
price_variation_by_company = car_data.groupby('Company')['Price ($)'].std().sort_values(ascending=False)

# Find the company with the highest price variation
highest_variation_company = price_variation_by_company.idxmax()
highest_variation_value = price_variation_by_company.max()

# Display the results
highest_variation_company, highest_variation_value, price_variation_by_company
```

```python
#Q-3 What is the distribution of car prices for each transmission type, and how do the interquartile ranges.Compare?
#ANS-
# Group the data by Transmission and calculate the distribution statistics for car prices
price_distribution_by_transmission = car_data.groupby('Transmission')['Price ($)'].describe()

# Extract the interquartile range (IQR) for each transmission type
price_distribution_by_transmission['IQR'] = (
    price_distribution_by_transmission['75%'] - price_distribution_by_transmission['25%']
)

# Display the distribution statistics and the IQR comparison
price_distribution_by_transmission[['25%', '50%', '75%', 'IQR']]
```

```python
#Q-12 Is there any seasonal trend in car sales based on the date of sale?
#ANS-
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset (replace this with your actual data)
# Let's assume the dataframe has a 'sale_date' and 'sales' column
data = {
    'sale_date': [
        '2023-01-15', '2023-03-22', '2023-06-10', '2023-08-05', '2023-11-19', '2023-12-25',
        '2023-02-13', '2023-04-17', '2023-07-08', '2023-10-30', '2023-09-05', '2023-05-14'
    ],
    'sales': [150, 200, 250, 210, 180, 300, 160, 220, 240, 210, 230, 190]
}

# Convert to a pandas DataFrame
df = pd.DataFrame(data)

# Convert 'sale_date' to datetime
df['sale_date'] = pd.to_datetime(df['sale_date'])

# Extract month and season (Winter, Spring, Summer, Fall)
df['month'] = df['sale_date'].dt.month
df['season'] = df['month'].apply(lambda x: 'Winter' if x in [12, 1, 2] else
                                          'Spring' if x in [3, 4, 5] else
                                          'Summer' if x in [6, 7, 8] else 'Fall')

# Group by month and season, summing the sales
monthly_sales = df.groupby('month')['sales'].sum()
seasonal_sales = df.groupby('season')['sales'].sum()

# Plotting monthly sales
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Car Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)

# Plotting seasonal sales
plt.subplot(1, 2, 2)
seasonal_sales.plot(kind='bar', color='lightcoral')
plt.title('Seasonal Car Sales')
plt.xlabel('Season')
plt.ylabel('Total Sales')

# Show the plots
plt.tight_layout()
plt.show()

```

```python
#Q-6 What are the most popular product keywords?
#ANS-

def extract_keywords(product_name):
  """Extracts keywords from a product name, handling potential numbers."""
  if isinstance(product_name, str):  # Check if it's a string
    keywords = product_name.lower().split()  # Split into words and lowercase
    keywords = [word for word in keywords if word.isalpha()]  # Remove non-alphabetical characters
  else:
    keywords = []  # Handle non-string values (e.g., integers) by returning an empty list
  return keywords

# Apply the function to extract keywords
df["keywords"] = df["product_name"].apply(extract_keywords)

# Flatten the list of keywords
all_keywords = [keyword for keywords in df["keywords"] for keyword in keywords]

# Count keyword occurrences
keyword_counts = pd.Series(all_keywords).value_counts()

# Display the top 10 most popular keywords
print(keyword_counts.head(10))
```

```python
#Q-2  What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram?
#ANS-
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/mnt/data/spotify.csv'
spotify_data = pd.read_csv('spotify.csv')

# Check for missing values in the 'popularity' column and drop them
spotify_data = spotify_data.dropna(subset=['Popularity'])

# Plot the histogram of popularity
plt.figure(figsize=(10, 6))
plt.hist(spotify_data['Popularity'], bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Popularity Among Tracks', fontsize=16)
plt.xlabel('Popularity', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Q-3  Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot?
#ANS-
import matplotlib.pyplot as plt

# Ensure there are no missing values in 'popularity' and 'duration_ms'
spotify_data_cleaned = spotify_data.dropna(subset=['Popularity', 'duration_ms'])

# Scatter plot to explore the relationship between popularity and duration
plt.figure(figsize=(10, 6))
plt.scatter(spotify_data_cleaned['Duration_ms'], spotify_data_cleaned['popularity'], alpha=0.5, color='blue', edgecolors='black')
plt.title('Relationship Between Track Duration and Popularity', fontsize=16)
plt.xlabel('Duration (ms)', fontsize=14)
plt.ylabel('Popularity', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

```python
#Q-4 Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot?
#ANS-
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/mnt/data/spotify.csv'
spotify_data = pd.read_csv('spotify.csv')

# Group by artist and count the number of tracks
artist_track_counts = spotify_data['Artist'].value_counts().reset_index()
artist_track_counts.columns = ['Artist', 'track_count']

# Display the top 10 artists with the most tracks
top_artists = artist_track_counts.head(10)

# Plot the count of tracks for each artist
plt.figure(figsize=(12, 6))
sns.barplot(data=top_artists, x='track_count', y='Artist', palette='viridis')
plt.title('Top 10 Artists with the Most Tracks', fontsize=16)
plt.xlabel('Number of Tracks', fontsize=14)
plt.ylabel('Artist', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q-5 What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each?
#ANS-
import pandas as pd

# Load the dataset
file_path = 'spotify.csv'  # Update with your local file path
spotify_data = pd.read_csv('spotify.csv')

# Ensure the 'popularity', 'track_name', and 'artist' columns are not null
spotify_data_cleaned = spotify_data.dropna(subset=['Popularity', 'Track Name', 'Artist'])

# Find the 5 least popular tracks
least_popular_tracks = spotify_data_cleaned.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']]

# Display the results
print("Top 5 Least Popular Tracks:")
print(least_popular_tracks)
```

```python
#Q-6 Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist?
#ANS-
import pandas as pd

# Load the dataset
file_path = 'spotify.csv'  # Replace with your local file path
spotify_data = pd.read_csv('spotify.csv')

# Ensure the 'popularity' and 'artist' columns are not null
spotify_data_cleaned = spotify_data.dropna(subset=['Popularity', 'Artist'])

# Calculate the average popularity for each artist
artist_popularity = spotify_data_cleaned.groupby('Artist')['Popularity'].mean().reset_index()

# Sort artists by average popularity in descending order and select the top 5
top_artists_by_popularity = artist_popularity.sort_values(by='Popularity', ascending=False).head(5)

# Display the results
print("Top 5 Artists by Average Popularity:")
print(top_artists_by_popularity)
```

```python
#Q-7  For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist?
#ANS-
import pandas as pd

# Load the dataset
file_path = 'spotify.csv'  # Update with your local file path
spotify_data = pd.read_csv('spotify.csv')

# Ensure the 'popularity', 'artist', and 'track_name' columns are not null
spotify_data_cleaned = spotify_data.dropna(subset=['Popularity', 'Artist', 'Track Name'])

# Identify the top 5 most popular artists based on average popularity
top_5_artists = spotify_data_cleaned.groupby('Artist')['Popularity'].mean().nlargest(5).index

# Filter the dataset for tracks by these top 5 artists
top_5_artist_tracks = spotify_data_cleaned[spotify_data_cleaned['Artist'].isin(top_5_artists)]

# Find the most popular track for each of these artists
most_popular_tracks = top_5_artist_tracks.loc[top_5_artist_tracks.groupby('Artist')['Popularity'].idxmax()][['Artist', 'Track Name', 'Popularity']]

# Display the results
print("Most Popular Tracks for the Top 5 Artists:")
print(most_popular_tracks)
```

```python
#Q-8 Visualize relationships between multiple numerical variables simultaneously using a pair plot?
#ANS-
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/mnt/data/spotify.csv'  # Update with the correct path if needed
spotify_data = pd.read_csv('spotify.csv')

# Select numerical columns for the pair plot
numerical_columns = ['popularity', 'duration_ms', 'tempo', 'danceability', 'energy']

# Ensure the selected columns do not have missing values
spotify_data_cleaned = spotify_data[numerical_columns].dropna()

# Create the pair plot
sns.pairplot(spotify_data_cleaned, diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle('Pair Plot of Numerical Features', y=1.02, fontsize=16)
plt.show()
```

```python
#Q-9 Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot?
#ANS
import seaborn as sns

import matplotlib.pyplot as plt

# Select relevant columns and drop missing values
artist_duration_data = spotify_data[['artist', 'duration_ms']].dropna()

# Filter to include only the top 10 artists by track count for better visualization
top_10_artists = artist_duration_data['artist'].value_counts().head(10).index
filtered_data = artist_duration_data[artist_duration_data['artist'].isin(top_10_artists)]

# Create a violin plot to visualize duration variation across top artists
plt.figure(figsize=(14, 8))
sns.violinplot(data=filtered_data, x='artist', y='duration_ms', palette='viridis')
plt.title('Variation in Track Duration Across Top 10 Artists', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Duration (ms)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q-10 How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot?
#ANS-
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'spotify.csv'  # Update with your local file path
spotify_data = pd.read_csv('spotify.csv' )

# Select relevant columns and drop missing values
artist_popularity_data = spotify_data[['artist', 'popularity']].dropna()

# Filter to include only the top 10 artists by track count for better visualization
top_10_artists = artist_popularity_data['artist'].value_counts().head(10).index
filtered_data = artist_popularity_data[artist_popularity_data['artist'].isin(top_10_artists)]

# Create a violin plot to visualize popularity distribution for top artists
plt.figure(figsize=(14, 8))
sns.violinplot(data=filtered_data, x='artist', y='popularity', palette='viridis', inner='quartile')
plt.title('Distribution of Track Popularity Across Top 10 Artists', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Popularity', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
